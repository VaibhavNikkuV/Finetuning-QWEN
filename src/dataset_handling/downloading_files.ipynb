```python
import os
import numpy as np
import pandas as pd
import requests
```

```python
file_path = "/mnt/e/FinetuningQwen/Finetuning/dataset/RAW/AWM_processed_data.csv"
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
def download_file(file_url, file_path):
    buffer_file = requests.get(file_url)
    with open(file_path, "wb") as f:
        f.write(buffer_file.content)
    return file_path
```

```python
main_folder = "/Users/d02505-vaibhav/Learning_Projects/Personal/Finetuning/dataset/RAW"

lang_dirs = {1: "English", 2: "Hindi"}
for sub in ("student", "teacher"):
    for lang in lang_dirs.values():
        os.makedirs(os.path.join(main_folder, sub, lang), exist_ok=True)

for idx, row in df.iterrows():
    lang = lang_dirs.get(row["language"], "Unknown")
    stem = str(row["id"])

    student_path = os.path.join(main_folder, "student", lang, f"{stem}.pdf")
    teacher_path = os.path.join(main_folder, "teacher", lang, f"{stem}.pdf")

    if not os.path.exists(student_path):
        print(f"Downloading student {stem} ({lang})")
        download_file(row["student_file_url"], student_path)
    if not os.path.exists(teacher_path):
        print(f"Downloading teacher {stem} ({lang})")
        download_file(row["teacher_file_url"], teacher_path)
```
